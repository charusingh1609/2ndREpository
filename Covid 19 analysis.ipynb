```python
def pipe(data):
    size = data.shape[0]
    
    for i in range(1,size):
        #rint(data.iloc[i])
        df = data.iloc[i]
        
        state = df[" State"]
        total_cases = df["Total Cases"]
        Active_cases =int(total_cases * df["Active Ratio (%)"] // 100 )
        Discharge_Ratio= int(total_cases* df["Discharge Ratio (%)"]//100)
        Death_Ratio=int(total_cases* df["Death Ratio (%)"]//100)
        
        print("State",state," Active cases", Active_cases, "Discharge cases ", Discharge_Ratio, "Death Ratio" , Death_Ratio)
        
        
        import matplotlib.pyplot as plt

# create data: an array of values
        names='Death_Ratio', ' Discharge_Ratio', ' Discharge_Ratio',
        values=[Death_Ratio, Discharge_Ratio,Active_cases]
        
        # Create a pieplot
        #plt.pie()
        plt.pie(values, labels=names, labeldistance=1.15);
        plt.title(state)
        plt.show()
        
        
        
    
```

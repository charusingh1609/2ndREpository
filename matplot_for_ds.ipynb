```python
plt.plot([1,2,3],[2,5,7])
plt.show()
```

```python
#how to add title
plt.plot([1,2,3],[2,5,7])
plt.title("dummy data")
plt.show()
```

```python
#how to add xlabel and y label
plt.plot([1,2,3],[2,5,7])
plt.title("dummy data")
plt.xlabel("x label")
plt.ylabel("y label")
plt.show()
```

```python
from matplotlib import pyplot as plt
x=[1,5,4]
y=[23,45,89]
x2=[3,7,3]
y2=[5,8,9]
plt.plot(x,y)
plt.plot(x2,y2)
plt.show()
```

```python
# FOR CHANGING STYLE
from matplotlib import pyplot as plt
from matplotlib import style
style.use('ggplot')
x=[1,5,4]
y=[23,45,89]
x2=[3,7,3]
y2=[5,8,9]
plt.plot(x,y)
plt.plot(x2,y2)
plt.show()
```

```python
# FOR giving label
from matplotlib import pyplot as plt
from matplotlib import style
style.use('ggplot')
x=[1,5,4]
y=[23,45,89]
x2=[3,7,3]
y2=[5,8,9]
plt.plot(x,y, label='first')
plt.plot(x2,y2, label='second')
plt.legend()
plt.show()
```

```python
# FOR giving linebidth
from matplotlib import pyplot as plt
from matplotlib import style
style.use('ggplot')
x=[1,5,4]
y=[23,45,89]
x2=[3,7,3]
y2=[5,8,9]
plt.plot(x,y, label='first', linewidth=6)
plt.plot(x2,y2, label='second')
plt.legend()
plt.show()
```

```python
# FOR giving linebidth
from matplotlib import pyplot as plt
from matplotlib import style
style.use('ggplot')
x=[1,5,4]
y=[23,45,89]
x2=[3,7,3]
y2=[5,8,9]
plt.plot(x,y, label='first', linewidth=6)
plt.plot(x2,y2, label='second')
plt.legend()
plt.show()
```

```python
# FOR USING SCATTERPLOT
from matplotlib import pyplot as plt
from matplotlib import style
style.use('ggplot')
x=[1,5,4]
y=[23,45,89]
x2=[3,7,3]
y2=[5,8,9]
plt.scatter(x,y, label='first', linewidth=6)
plt.plot(x2,y2, label='second')
plt.legend()
plt.show()
```

```python
# FOR using histogram
from matplotlib import pyplot as plt
from matplotlib import style
style.use('ggplot')
x=[1,5,4]
y=[23,45,89]
x2=[3,7,3]
y2=[5,8,9]
plt.hist(x, label='first', linewidth=6)
plt.hist(x2,y2, label='second')
plt.legend()
plt.show()
```
